```python
import os
import pandas as pd
from sqlalchemy import create_engine
```

```python
dictionary_file = "Resources/dictionary.csv"
summer_file = "Resources/summer.csv"
winter_file = "Resources/winter.csv"
```

```python
dictionary_df = pd.read_csv(dictionary_file)
summer_df = pd.read_csv(summer_file)
winter_df = pd.read_csv(winter_file)
```

```python
dictionary_df.head()
```

```python
country_codes_df = dictionary_df[['Country', 'Code']]
renamed_country_codes_df = country_codes_df.rename(columns={'Country':'country', 'Code':'code'})
renamed_country_codes_df
```

```python
summer_df.head()
```

```python
winter_df.head()
```

```python
usa_code = dictionary_df.loc[dictionary_df['Country'] == "United States"]
usa_code
```

```python
usa_summer_df = summer_df.loc[((summer_df['Country'] == 'USA') 
               & (summer_df['Medal'] == 'Gold')), 
            ['Year', 'Athlete', 'Country', 'Gender', 'Sport', 'Event']]
usa_summer_df
```

```python
usa_winter_df = winter_df.loc[((winter_df['Country'] == 'USA') 
               & (winter_df['Medal'] == 'Gold')), 
            ['Year', 'Athlete', 'Country', 'Gender', 'Sport', 'Event']]
usa_winter_df
```

```python
combined_df = usa_summer_df.append(usa_winter_df, ignore_index=True)
renamed_combined_df = combined_df.rename(columns={
    'Year':'year',
    'Athlete':'athlete',
    'Country':'country_code',
    'Gender':'gender',
    'Sport':'sport',
    'Event':'event'
})
renamed_combined_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/olympics_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
renamed_country_codes_df.to_sql(name='country_info', con=engine, if_exists='append', index=False)
```

```python
renamed_combined_df.to_sql(name='usa_golds', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from country_info', con=engine).head()
```
